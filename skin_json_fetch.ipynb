```python
import requests
import json
import pandas as pd
import time
```

```python
def fetch_price_history(hash_name):
    url = f"https://steamcommunity.com/market/pricehistory/?currency=1&appid=730&market_hash_name={hash_name}"
    headers = {
        "cookie": "Your cookie here"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to fetch data. Status code:", response.status_code)
        return None

def main():
    # Example DataFrame with skin names
    df = pd.read_csv('Datasets/skins_df.csv')

    for index, row in df.iloc[3977:,:].iterrows():
        skin_name = row['hash_name']
        #print(skin_name)
        price_history = fetch_price_history(skin_name)
        if price_history:
            with open(f"Jsons/skin_json/{skin_name}.json", "w") as f:
                json.dump(price_history, f, indent=4)
            #print(f"Price history for {skin_name} fetched and stored in {skin_name}.json")
        else:
            print(f"Failed to fetch price history for {skin_name}.")

        time.sleep(2)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import json
import os
```

```python
def parse_json(json_file):
    with open(json_file, 'r') as f:
        data = json.load(f)

    prices_data = data['prices']

    df = pd.DataFrame(prices_data, columns=['date', 'price', 'quantity'])

    df['date'] = pd.to_datetime(df['date'], format='%b %d %Y %H: +%f')
    df['hash_name'] = json_file[:-5]

    return df
```

# Creating Dataframe of all json files

```python
path_to_json = 'Jsons/'
```

```python
# Initialize a list to store JSON file paths
json_files = []

# Recursively search for JSON files within the JSON directory
for root, dirs, files in os.walk(os.path.join(path_to_json)):
    for file in files:
        if file.endswith('.json'):
            json_files.append(os.path.join(root, file))

# Print the list of JSON file paths
for file in json_files:
    if "ipynb" in file:
        print(file)
        
print(json_files[:5])
```

```python
all_data = pd.DataFrame()

for file in json_files:
    df = parse_json(file)
    all_data = pd.concat([all_data, df], ignore_index=True)

# all_data = all_data.sort_values(by='date')

all_data.reset_index(drop=True, inplace=True)

all_data
```

```python
all_data.to_csv('Datasets/all_data.csv', index = False)
```
